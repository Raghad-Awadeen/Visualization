#### 1. Using a line plot visualize the Total profit data provided for each month.

**Note:**It must be implemented in three methods
- Generated line plot must include the following properties:

X axis name = Month Number <br>
Y axis name = Total profit <br>
The line plot graph should look like this.


```python
#Q1
#(Line Plot using Matplotlib)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
company=pd.read_csv('Data/company_sales_data.csv')
company.head()
profitList = company ['total_profit'].tolist()
monthList  = company ['month_number'].tolist()
plt.plot(monthList, profitList, label = 'Month-wise Profit data of last year')
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.xticks(monthList)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
#-------------------------------------------------------------------------------------------------------------

#---------------------------------------------------------------------------------------------------------------
#(Line Plot using Seaborn)
#company_d=company.drop(['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer','total_units'], axis=1)
sns.lineplot( data=company,x='month_number',y='total_profit' )
```

```python
#(Line Plot using Pandas visualization)
profitList = company ['total_profit'].tolist()
monthList  = company ['month_number'].tolist()
df_company=pd.DataFrame({'month': monthList ,'total':profitList })
ax = plt.gca()
df_company.plot( x= 'month', y= 'total', ax = ax)
plt.xlabel('Month')
plt.ylabel('Total profit')
plt.title('Company profit per month')
plt.legend()

plt.show()
```

```python
#Q2
#(multiline plot using Matplotlib)
company.describe()
profitList = company ['total_profit'].tolist()
monthList  = company ['month_number'].tolist()
facecreamList  = company ['facecream'].tolist()
facewashList  = company ['facewash'].tolist()
toothpasteList  = company ['toothpaste'].tolist()
bathingsoapList  = company ['bathingsoap'].tolist()
shampooList  = company ['shampoo'].tolist()
moisturizerList  = company ['moisturizer'].tolist()

plt.plot(monthList, facecreamList,   label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(monthList, facewashList,   label = 'Face Wash Sales Data', marker='o', linewidth=3)
plt.plot(monthList, toothpasteList,   label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(monthList, bathingsoapList,   label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(monthList, shampooList,   label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(monthList, moisturizerList,   label = 'ToothPaste Sales Data', marker='o', linewidth=3)


plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.xticks(monthList)
plt.title('Company profit per month')
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.legend(loc='upper left')

plt.show()
```
